```python
test_spec = pd.read_csv('TestSpectra.csv')
Predictions['Places'] = test_spec.iloc[:,0].tolist()
df = test_spec.drop(columns=test_spec.columns[0]) 
```

```python
test_spectra = df.iloc[:, 51::].copy()
test_spectra.head(2)
```

```python
spec2t = {}
for i in window_lengths:
    spec2t[i] = filt_sg(test_spectra, i, 'sg2')

smth_spect = sgsmooth (test_spectra, 3)    
```

```python
cr_spect = continuum_removed(spec2t[51])
log_spect = ((1/spec2t[51]).apply(np.log)).copy()
```

```python
bands_sampling = [7, 21, 27, 37, 50]
sampled_spect = {}
sampled_crt = {}
fod_sampledt = {}
sampled_logt = {}
for n in bands_sampling:
    sampled_spect[n] = resample_spectra (spec2t[51], n)
    sampled_crt[n] = resample_spectra (cr_spect, n)
    fod_sampledt[n] = fod (sampled_spect[n]) 
    sampled_logt[n] = resample_spectra (log_spect, n)
```

## Prediction for Sand, Silt, Clay TOC and CaCO3 using the best Models.

### Cubist (sand)

#### Leave one out

Sand -> [ r2 - **0.59**, Spec - log, n_bands -7]     <br/><br/>
***Description***: To obtain the best model for sand using cubist method we trained the model on the logarithmic of the reflectance inverse spectra which is  resampled to 7 bands.

```python
sand_model_cubist = Cubist( n_committees = 10, n_rules = 10 ) 
sand_model_cubist.fit(sampled_log[7],  T[0])
```

```python
Predictions['Sand'] = sand_model_cubist.predict(sampled_logt[7])
```
