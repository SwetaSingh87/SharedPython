### GBRT

Sand -> [ r2 - **0.38**, Spec - cr, n_bands - 10]   <br/><br/>

 continuum removal of the smoothed spectra  which was resampled to the 10 bands provided by the Sentinel.


```python
# Check the shape of the DataFrame
print("Shape of sampled_cr10:", sampled_cr10.shape)

# Check the data types of each column
print("Data types of sampled_cr10:\n", sampled_cr10.dtypes)

# Check for missing values
print("Missing values in sampled_cr10:\n", sampled_cr10.isnull().sum())

# Display the first few rows of the DataFrame
print("First few rows of sampled_cr10:\n", sampled_cr10.head())
```

```python
sand_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 5, n_estimators = 20, random_state = 42) 
sand_model_gbrt.fit(sampled_cr10_copy, T0_copy)
```
