##  Sand 

### GBRT

Sand -> [ r2 - **0.32**, Spec - fod, n_bands -11]     <br/><br/>
***Description***: To obtain the best model for sand using gbrt method we trained the model on the first order derivative of the  smoothed spectra which was  resampled to the 11 bands provided by Venus.

```python
# Check the shape of the DataFrame
print("Shape of fod_sampled11:", fod_sampled11.shape)

# Check the data types of each column
print("Data types of fod_sampled11:\n", fod_sampled11.dtypes)

# Check for missing values
print("Missing values in fod_sampled11:\n", fod_sampled11.isnull().sum())

# Display the first few rows of the DataFrame
print("First few rows of fod_sampled11:\n", fod_sampled11.head())
```

```python
fod_sampled11_copy = fod_sampled11.copy()
T0_copy = T[0].copy()
```

```python
# Check the types
print("Type of fod_sampled11_copy:", type(fod_sampled11_copy))
print("Type of T0_copy:", type(T0_copy))
```

```python
sand_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 5, n_estimators = 40, random_state = 42) 
sand_model_gbrt.fit(fod_sampled11_copy,  T0_copy)
```

```python
feature_scores['Sand'] = permutation_importance(sand_model_gbrt, fod_sampled11_copy, T0_copy, n_repeats=20, random_state=0)
model_details['Sand'] = {'method': 'gbrt', 'spec': fod_sampled11, 'spec_proc': 'fod', 'n_bands': 11}
```

```python
show_important_features_for ('Sand', model_details, feature_scores, 5, 40, 2)
```
