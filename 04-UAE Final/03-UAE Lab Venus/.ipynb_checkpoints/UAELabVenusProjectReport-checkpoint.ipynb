#  ***Report of UAE Lab Data Resampled to Venus (11 bands)***

### Noise Removal and Spectral Preprocessing

1. Noise Removal and Smoothing of Spectra :- 
    - Noise is removed from the spectras by removing first 50 wavelengths (350 nm - 400 nm). 
    - Then the spectras are smoothed using the 2nd order savitzky golay smoothing filter applied at the window length of 51.

The following spectral preprocessing options are considered on the smoothed spectras :
   1. None of the above.
   2. Continuum Removal.
   3. Logarithmic transformation on the inverse of the spectra.

Each of the three different kinds of spectra are resampled to the 11 bands provided by Venus. For all the resampled spectra corresponding to the three different preprocessing, we also consider first order derivative of the resampled smoothed spectra, resampled continuum removed spectra and resampled logarithmic transformed spectra. Thus finally resulting in the following six categories:
- *sampled_spec11* : smoothed spectra resampled to the 11 bands provided by Venus.
- *sampled_cr11* :   continuum removed spectra resampled to  the 11 bands provided by Venus.
- *sampled_log11* : logarithmic transformed spectra resampled to  the 11 bands provided by Venus.
- *fod_sampled11* : first order derivative of the smoothed spectra resampled to the 11 bands provided by Venus.
- *fod_cr11* : first order derivative of the continuum removed spectra resampled to the 11 bands provided by Venus.
- *fod_log11* : first order derivative of the logarithmic transformed spectra resampled to the 11 bands provided by Venus.

```python
data = {'Continuum': sampled_cr11.iloc[0,:].T, 'Log': sampled_log11.iloc[0,:].T, \
        'None': sampled_spec11.iloc[0,:].T,  'fod': 100*fod_sampled11.iloc[21,:].T, \
        'fod_cr': fod_cr11.iloc[0,:].T,  'fod_log': fod_log11.iloc[0,:].T}
dfplot = pd.DataFrame(data)

#define list of subplot titles
title_list = ['Continuum', 'Log', 'None', 'fod', 'fod_cr', 'fod_log']
dfplot.plot(kind='line', subplots=True, title=title_list, grid=False,
        layout=(3, 2), sharex=True, sharey=False, 
        style=['g', 'y', 'b', 'r', 'g', 'b'],)

for ax in plt.gcf().axes:
    ax.get_legend().remove()

    
plt.suptitle("Different Kinds of Spectral Preprocessing")
plt.tight_layout()
plt.show()
       
```

### Obtaining Hypertuned Models for Each  Preprocessed Spectra

- For each preprocessed spectra and a specific machine learning algorithm a hypertuned model was obtained using leave one out     method.
- The model accuracy of each hypertuned model was evaluated using leave one out method. 
- Six different machine learning algorithms used in the data analysis are :
   1. Multile Linear Regression as '**mult**'.
   2. Partial Least Square Regression as '**PLSR**'.
   3. Kernel Ridge Regression as '**ridge**'.
   4. Support Vector Regression as '**SVR**'.
   5. Cubist as '**Cubist**'.
   6. Gradient Boosting Regression tree as '**GBRT**'.

##  Sand

### Gradient Boosting Regression Tree 

Sand -> [ r2 - **0.32**, Spec - fod, n_bands -11]     <br/><br/>
***Description***: To obtain the best model for sand using gbrt method we trained the model on the first order derivative of the  smoothed spectra which was  resampled to the 11 bands provided by Venus.

```python
sand_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 5, n_estimators = 40, random_state = 42) 
sand_model_gbrt.fit(fod_sampled11,  T[0])
```

```python
feature_scores['Sand'] = permutation_importance(sand_model_gbrt, fod_sampled11, T[0], n_repeats=20, random_state=0)
model_details['Sand'] = {'method': 'gbrt', 'spec': fod_sampled11, 'spec_proc': 'fod', 'n_bands': 11}
show_important_features_for ('Sand', model_details, feature_scores, 5, 50, 20)
```
