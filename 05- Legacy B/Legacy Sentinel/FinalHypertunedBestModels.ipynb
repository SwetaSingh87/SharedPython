 #  Best Models for Sand, Silt, Clay, TOC and CaCO3.

##  Sand 

### Cubist

 Sand -> [ r2 - **0.60**, Spec - fod_log, n_bands -10]   <br/><br/>
***Description***:To obtain the best model for sand using cubist method we trained the model on the first order derivative of the logarithmic transformed  spectra which was resampled to the 10 bands provided by Sentinel.


```python
sand_model_cubist = Cubist( n_committees = 15, n_rules = 10) 
sand_model_cubist.fit(fod_log10,  T[0])
```

```python
with open('sand_model_cubist.pkl', 'wb') as f:
    pickle.dump(sand_model_cubist, f)
```

## Silt

### Gradient Boosting Regression Tree

Silt -> [ r2 - **0.41**, Spec - fod, n_bands - 10]   <br/><br/>
***Description***:To obtain the best model for silt using gradient boosting regression tree method we trained the model on the first order derivative of the smoothed spectra which was resampled to the 10 bands provided by the Sentinel.


```python
silt_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 3, n_estimators = 50, random_state = 42)
silt_model_gbrt.fit(fod_sampled10,  T[1])
```

```python
with open('silt_model_gbrt.pkl', 'wb') as f:
    pickle.dump(silt_model_gbrt, f)
```
