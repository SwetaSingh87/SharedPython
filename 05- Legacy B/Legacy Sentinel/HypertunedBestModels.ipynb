##  Sand 

### Cubist

 Sand -> [ r2 - **0.60**, Spec - fod_log, n_bands -10]   <br/><br/>
***Description***:To obtain the best model for sand using cubist method we trained the model on the first order derivative of the logarithmic transformed  spectra which was resampled to the 10 bands provided by Sentinel.


```python
sand_model_cubist = Cubist( n_committees = 15, n_rules = 10) 
sand_model_cubist.fit(fod_log10,  T[0])
```

```python
feature_scores['Sand'] = permutation_importance(sand_model_cubist, fod_log10, T[0], n_repeats=20, random_state=0)
model_details['Sand'] = {'method': 'cubist', 'spec': fod_log10, 'spec_proc': 'fod_log', 'n_bands': 10}
```

```python
show_important_features_for ('Sand', model_details, feature_scores, 2, 10, 2)
```

### Gradient Boosting Regression Tree

Sand -> [ r2 - **0.56**, Spec - fod_log, n_bands - 10]   <br/><br/>

```python
sand_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 4, n_estimators = 20, random_state = 42)
sand_model_gbrt.fit(fod_log10,  T[0])
```
