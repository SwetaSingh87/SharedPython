##  Sand 

### GBRT

Sand -> [ r2 - **0.49**, Spec - fod2, n_bands - 60]   <br/><br/>
***Description***:To obtain the best model for sand using gbrt method we trained the model on the
 first order derivative of the smoothed spectra  which is resampled to 60 bands.


```python
sand_model_gbrt = GradientBoostingRegressor(learning_rate = 0.1, max_depth = 3, n_estimators = 30, random_state = 42) 
sand_model_gbrt.fit(fod_sampled[60], T[0])
```
